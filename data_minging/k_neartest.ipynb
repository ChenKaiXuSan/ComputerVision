```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

# import dateset 
url = "iris_noise.csv"

x_dataset = pd.read_csv(url, usecols=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

y_dataset = pd.read_csv(url, usecols=['class'])

x_dataset_array = x_dataset.to_numpy()
y_dataset_array = y_dataset.to_numpy()
```

```python
# take model
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_dataset_array, y_dataset_array, test_size = 30/170, random_state=0)
```

```python
# test k in loop
# weights = uniform
acc = []

# Calculating error for K values between 1 and 40
for i in range(1, 100):
    knn = KNeighborsClassifier(n_neighbors=i, weights='uniform')
    knn.fit(x_train, y_train)
    assess_model_socre=knn.score(x_test,y_test)

    acc.append(assess_model_socre)
```

```python
# make figure
plt.figure(figsize=(12, 6))
plt.plot(range(1, 100), acc, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Acc K Value when weights is uniform')
plt.xlabel('K Value')
plt.ylabel('acc')
```

```python
# test k in loop 
# weights='distance'
acc = []

# Calculating error for K values between 1 and 40
for i in range(1, 100):
    knn = KNeighborsClassifier(n_neighbors=i, weights='distance')
    knn.fit(x_train, y_train)
    assess_model_socre=knn.score(x_test,y_test)

    acc.append(assess_model_socre)
```

```python
# make figure
plt.figure(figsize=(12, 6))
plt.plot(range(1, 100), acc, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Acc K Value when weights is distance')
plt.xlabel('K Value')
plt.ylabel('acc')
```
